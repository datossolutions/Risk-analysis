```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Master_Data_Sanitized_09032019.xlsx", sheet_name="Analyzed Data")
df
```

```python
cf = pd.read_excel("Outbound_File_Dario_16032019.xlsx", sheet_name="CF") 
cf_seller = pd.read_excel("Outbound_File_Dario_16032019.xlsx", sheet_name="CFSellers") 
#CF significa counterfeit que quiere decir que son productos falsos
```

```python
df["Days"] = df["Reseller Sale Date"] - df["Reported Date"] 
```

```python
#Regla 1 campos de Cisco tengan datos incompletos(si hay nans = 16 puntos y no pasar por las demás reglas)

validity= df["Validity"].tolist()
dates = df["Reported Date"].tolist()
rule_1=np.zeros(df.shape[0])
for i in range(len(validity)):
    if validity[i]== np.nan and dates[i]== np.nan :
        rule_1[i]=16
    else:
        rule_1[i]=0

            
df["Rule_1"]=rule_1            
```

```python
# Regla 2: Comparación entre el PID original y el Reseller PID. 
df['Rule_2'] = np.where((df['PID'] == df['Reseller PID']), 0, 6)
    
```

```python
#Regla 3: Chequear si el Reseller PID se encuentra en la lista adjunta (tab llamado CF), si esta en la lista se asignaran 3 puntos.

rule_3 =np.zeros(df.shape[0])
comparative = df["Reseller PID"].tolist()
cflist=cf.CF.tolist()
for i in range(len(comparative)):
    if comparative[i] in cflist:
        rule_3[i]= 3 #Sí está clasificada como counterfeit
    else:
        rule_3[i]= 0 #No está clasificada como counterfeit
        
        
df["Rule_3"]=rule_3
```

```python
net_price =  df['Cisco Total Net Price'].tolist()
reseller_price = df['Reseller Price'].tolist()
Rule_4 = np.zeros(df.shape[0])
days = df.Days.astype(int).tolist()

for i in range(len(Rule_4)):
    if reseller_price[i] > net_price[i]:
        Rule_4[i]=0
    elif reseller_price[i] < net_price[i]*0.95:
        Rule_4[i]=13
    elif(days[i] <= 90 and net_price[i]*0.95 <= reseller_price[i] <= net_price[i]):       
        Rule_4[i]=0
    else:
        Rule_4[i]=13
        
df["Rule_4"] = Rule_4         
```

```python
discount = df['Cisco Net Discount'].tolist()
net=df['Cisco Total Net Price'].tolist()
reseller_price = df['Reseller Price'].tolist()
days = df.Days.astype(int).tolist()
Rule_5 = np.zeros(df.shape[0])
for i in range(len(Rule_5)):
    if discount[i] > 0.42 and days[i] < 180 and reseller_price[i] < net[i]:
        Rule_5[i] = -6
    else:
        Rule_5[i] = 0

df["Rule_5"]=Rule_5    
```

```python
#Regla 6:
#Chequear si alguna de las entidades envueltas en la transacción, 
#también están presentes en la lista del tab CFSeller en el archivo adjunta, si estan en la lista se asignaran 6 puntos.

rule_6 =np.zeros(df.shape[0])
comparative = df['Reseller Vendor'].tolist()
cf_sellerlist=cf_seller.CFSellers.tolist()
for i in range(len(comparative)):
    if comparative[i] in cf_sellerlist:
        rule_6[i]=6 #Sí está clasificada como counterfeit
    else:
        rule_6[i]=0 #No está clasificada como counterfeit
        
        
df["Rule_6"]=rule_6
```

```python
#Regla 7:Si el Reseller Vendor se vio involucrado en la venta inicial del producto, 
#es decir, que aparezca en el campo Reseller o en el campo Reported End User, entonces se asignaran -10 puntos.

rule_7 =np.zeros(df.shape[0])
Reseller_vendor = df['Reseller Vendor'].tolist()
Reprorted_end_user= df['Reported End User'].tolist()
Reseller=df.Reseller.tolist()
for i in range(len(Reseller_vendor)):
    if Reseller_vendor[i] == Reprorted_end_user[i]:
        rule_7[i]= -10 #Sí está clasificada como counterfeit
        
    elif Reseller_vendor[i] == Reseller[i] :
        
        rule_7[i]= -10 #Sí está clasificada como counterfeit
        
    else:
        rule_7[i]= 0
df["Rule_7"]=rule_7
```

```python
#Regla 8: si el último vendedor tiene un SmartNet Contract para el producto.
#If original end customer has a SMARTnet contract for the product, then 3 points
#If not, then 0 points

        
df["Rule_8"]=df['SMARTnet and End Customer'].isna()

Rule_8 = df["Rule_8"].tolist()
ver=np.zeros(df.shape[0])
for i in range(len(ver)):
    if Rule_8[i] == False:
        ver[i]=3
    else:
        ver[i]=0
        
        
df["Rule_8"]=ver 
```

```python
df["Risk"]= df[['Rule_1', 'Rule_2', 'Rule_3', 'Rule_4',"Rule_5", 'Rule_6','Rule_7','Rule_8']].T.sum()
```

```python
Risk = df["Risk"].tolist()
ver=[]

for i in range(len(Risk)):
    if Risk[i] > 15:
        ver.append("HIGH RISK COUNTERFEIT")
    elif 6 < Risk[i] <15:
        ver.append("MEDIUM RISK COUNTERFEIT")
    else:
        ver.append("LOW RISK COUNTERFEIT")

df["Risk_Category"] = ver
```
